```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib

sns.set(style='whitegrid')
RANDOM_STATE = 42
```

```python

DATA_PATH = "iris.csv"   
ZIP_PATH = r"C:\Users\manoj\Downloads\dataset1.zip"  
if os.path.exists(DATA_PATH):
    df = pd.read_csv(DATA_PATH)
    print(f"Loaded CSV from {DATA_PATH}")
else:
    if os.path.exists(ZIP_PATH):
        import zipfile
        with zipfile.ZipFile(ZIP_PATH, 'r') as z:
            names = z.namelist()
            print("ZIP contents:", names)
            # try to find first .csv
            csvs = [n for n in names if n.lower().endswith('.csv')]
            if not csvs:
                raise FileNotFoundError("No CSV file found inside the zip.")
            extract_dir = "extracted_data"
            z.extractall(extract_dir)
            csv_path = os.path.join(extract_dir, csvs[0])
            df = pd.read_csv(csv_path)
            print(f"Extracted and loaded {csvs[0]} from zip.")
    else:
        raise FileNotFoundError("Neither DATA_PATH nor ZIP_PATH found. Edit DATA_PATH or ZIP_PATH in the cell.")

# show basic info
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
df.info()
display(df.describe(include='all'))
print("Value counts of species (if exists):")
if 'species' in df.columns:
    print(df['species'].value_counts())
else:
    print("No 'species' column found; check your CSV column names.")
```

```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
label_col = 'species'  

missing = [c for c in feature_cols + [label_col] if c not in df.columns]
if missing:
    raise KeyError(f"Expected column(s) not found in DataFrame: {missing}")

X_df = df[feature_cols].copy()
y_series = df[label_col].copy()

print("Feature dtypes:\n", X_df.dtypes)
X_df = X_df.apply(pd.to_numeric, errors='raise')

le = LabelEncoder()
y_encoded = le.fit_transform(y_series)
print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))

X = X_df.values
y = y_encoded
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y)

print("Shapes => X_train:", X_train.shape, "X_test:", X_test.shape)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
print("Scaling complete. Example scaled row:", X_train_scaled[0])
```

```python
dt = DecisionTreeClassifier(random_state=RANDOM_STATE)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
print("Decision Tree accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt, target_names=le.classes_))
```

```python
plt.figure(figsize=(12,8))
plot_tree(
    dt,
    feature_names=feature_cols,
    class_names=le.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree")
plt.show()

# Save PNG
fig = plt.gcf()
fig.savefig("decision_tree.png", dpi=300, bbox_inches='tight')
print("Saved decision_tree.png")
```

```python
cm = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('DT Confusion Matrix')
plt.show()
```

```python
lr = LogisticRegression(max_iter=500, multi_class='multinomial', solver='lbfgs', random_state=RANDOM_STATE)
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)
print("Logistic Regression accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr, target_names=le.classes_))

cm_lr = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(6,4))
sns.heatmap(cm_lr, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('LR Confusion Matrix')
plt.show()
```

```python
cv_dt = cross_val_score(DecisionTreeClassifier(random_state=RANDOM_STATE), X, y, cv=5)
cv_lr = cross_val_score(LogisticRegression(max_iter=500, multi_class='multinomial', solver='lbfgs'),
                        scaler.fit_transform(X), y, cv=5)

print("DT CV scores:", np.round(cv_dt,3), "mean:", np.round(cv_dt.mean(),3))
print("LR CV scores:", np.round(cv_lr,3), "mean:", np.round(cv_lr.mean(),3))
```

```python
param_grid = {
    'max_depth': [None, 2, 3, 4, 5],
    'min_samples_split': [2, 3, 4],
    'criterion': ['gini', 'entropy']
}
gs = GridSearchCV(DecisionTreeClassifier(random_state=RANDOM_STATE), param_grid, cv=5, n_jobs=-1)
gs.fit(X_train, y_train)
print("Best DT params:", gs.best_params_)
best_dt = gs.best_estimator_
print("Best DT CV score on train:", gs.best_score_)
# Evaluate best_dt on test
y_pred_best_dt = best_dt.predict(X_test)
print("Best DT Test accuracy:", accuracy_score(y_test, y_pred_best_dt))
print(classification_report(y_test, y_pred_best_dt, target_names=le.classes_))
```

```python
joblib.dump(best_dt, "iris_decision_tree_best.pkl")
joblib.dump(lr, "iris_logistic_regression.pkl")
joblib.dump(scaler, "iris_scaler.pkl")
print("Saved models and scaler to current directory.")
```

```python
def predict_species(model, scaler_obj, sample):
    """
    model: trained sklearn model (dt or lr)
    scaler_obj: scaler (StandardScaler) used for training (use None for tree-only if you prefer)
    sample: list/array with 4 numeric values in order [sepal_length,sepal_width,petal_length,petal_width]
    """
    sample = np.array(sample).reshape(1, -1)
    if scaler_obj is not None:
        # scale first; scaling a tree input is harmless
        sample = scaler_obj.transform(sample)
    pred = model.predict(sample)   # returns encoded label
    return le.inverse_transform(pred)[0]

# Examples:
print("DT predicts:", predict_species(best_dt, scaler, [5.1, 3.5, 1.4, 0.2]))
print("LR predicts:", predict_species(lr, scaler, [5.1, 3.5, 1.4, 0.2]))
```
